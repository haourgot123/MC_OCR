```python
import re

# Chuỗi chứa ngày tháng cần phát hiện
text = """
Hôm nay là 22/10/2024. Ngày sinh của tôi là 01-01-1990.
Một số ngày khác: 2024/10/21, 15/06/2020, 07-07-2023, 28/10.
"""

# Regex để phát hiện các ngày tháng
date_pattern = r'(\b\d{2}[/-]\d{2}[/-]\d{4}\b)|(\b\d{4}[/-]\d{2}[/-]\d{2}\b)'

# Tìm tất cả các chuỗi ngày tháng phù hợp với regex
dates = re.findall(date_pattern, text)

# Kết quả sẽ trả về danh sách các tuple, cần lọc lại các giá trị phù hợp
extracted_dates = [date[0] or date[1] for date in dates]

# In ra danh sách ngày tháng tìm được
print(extracted_dates)
```

```python
import re

# Chuỗi chứa ngày tháng và thời gian cần phát hiện
text = """
Hôm nay là 22/10/2024. Cuộc họp diễn ra vào lúc 14:30:15.
Ngày sinh của tôi là 01-01-1990 và tôi sẽ gặp bạn vào lúc 07:45.
Một số ngày khác: 2024/10/21, 15/06/2020. Tôi thường thức dậy lúc 05:15:30 và ăn sáng lúc 17:40.
Hẹn gặp bạn lúc 10.31 hoặc 10 giờ 31 phút.
"""

# Regex để phát hiện ngày tháng (dd/mm/yyyy, mm-dd-yyyy, yyyy/mm/dd)
date_pattern = r'\b\d{2}[/-]\d{2}[/-]\d{4}\b|\b\d{4}[/-]\d{2}[/-]\d{2}\b'

# Regex để phát hiện thời gian (hh:mm:ss, hh:mm, hh.mm, hh giờ mm phút)
time_pattern = r'\b([01]?\d|2[0-3]):[0-5]\d(:[0-5]\d)?\b|\b([01]?\d|2[0-3])\.[0-5]\d\b'

# Kết hợp cả hai biểu thức
combined_pattern = f'({date_pattern})|({time_pattern})'

# Tìm tất cả các chuỗi ngày tháng và thời gian
matches = re.findall(combined_pattern, text)

# Lọc và gom kết quả vào một danh sách (chỉ lấy giá trị không rỗng)
extracted_info = [match[0] or match[1] for match in matches if match[0] or match[1]]

# In ra danh sách ngày tháng và thời gian tìm được
print(extracted_info)
```

```python
import numpy as np
import pandas as pd


def regex(text : str) -> bool:
    time_corpus = ['Ngày','NGÀY', 'GIỜ', 'Giờ', 'ngày', 'giờ', 'Gio', 'THỜI GIAN', 'Thời gian', 'thời gian','PRINT TIME', 'Print time','print time', 'NGÀY KINH DOANH', 'Date', 'DATE', 'Order']
    date_pattern = r'\b\d{2}[/-]\d{2}[/-]\d{4}\b|\b\d{4}[/-]\d{2}[/-]\d{2}\b'
    time_pattern = r'\b([01]?\d|2[0-3]):[0-5]\d(:[0-5]\d)?\b|\b([01]?\d|2[0-3])\.[0-5]\d\b'
    combined_pattern = f'({date_pattern})|({time_pattern})'
    matches = re.findall(combined_pattern, text)
    if matches:
        return True
    else:
        for corpus in time_corpus:
            match = re.search(r'\b' + re.escape(corpus) + r'\b', text, re.IGNORECASE)
            if match:
                return True
        return False



```
